```
# For training only
path = '/content/drive/My Drive/Colab Notebooks/subimg_pair.npy'
subimg_pair = np.load(path,allow_pickle = True)[()]

images = []
masks = []
ignores = []

for val in subimg_pair.values():
  img, mask, ign = val
  images.append(img)
  mask = np.ones_like(mask) - mask
  masks.append(mask)
  ignores.append(ign)

images = np.array(images)
masks = np.array(masks)
ignores = np.array(ignores)

del subimg_pair
```

```
from deepCR import train
```

```
num = 9984
# del trainer
trainer = train(images[:num,:,:], masks[:num,:,:], name='norm_true_90',norm = True, percentile_limit = 90, epoch=50,
                save_after=20, plot_every=10, use_tqdm=False)
```

```
trainer.train()
```

```

np.save(path.replace('subimg_pair.npy','norm_true_0.9_loss.npy'),np.array(trainer.validation_loss))
```
